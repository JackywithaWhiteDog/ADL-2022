```python
num_epochs = 3
epoch_steps = len(results['loss']) // num_epochs
plot_epoch = 100

plt.figure(dpi=100)
plt.plot([results['loss'][i] for i in range(0, len(results['loss']), plot_epoch)])
plt.xticks([i*epoch_steps/plot_epoch for i in range(num_epochs+1)], range(num_epochs+1))
plt.xlabel('Epoch')
plt.ylabel('RL loss')
plt.show()
```

```python
metrics = {
    'Precision': 'p',
    'Recall': 'r',
    'F1 score': 'f'
}
rouges = ['rouge-1', 'rouge-2', 'rouge-l']
epochs = range(1, len(results['rouge'])+1)

for m, k in metrics.items():
    plt.figure(dpi=100)
    for r in rouges:
        plt.plot(epochs, [elem[r][k] for elem in results['rouge']])
    plt.title(m)
    plt.legend(rouges)
    plt.xticks(epochs)
    plt.xlabel('Epoch')
    plt.ylabel(m)
    plt.show()
```
