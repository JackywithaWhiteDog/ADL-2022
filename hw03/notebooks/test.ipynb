```python
import json

from tw_rouge import get_rouge
```

```python
pred = []
with open("../pred/public.jsonl", "r") as f:
    for line in f:
        pred.append(json.loads(line))
len(pred), pred[0].keys()
```

```python
label = []
with open("../data/public.jsonl", "r") as f:
    for line in f:
        label.append(json.loads(line))
len(label), label[0].keys()
```

```python
decoded_label = []
with open("../label.jsonl", "r") as f:
    for line in f:
        decoded_label.append(json.loads(line))
len(decoded_label), decoded_label[0].keys()
```

```python
pred = sorted(pred, key=lambda x: x["id"])
label = sorted(label, key=lambda x: x["id"])
decoded_label = sorted(decoded_label, key=lambda x: x["id"])
```

```python
for p, l in zip(pred, label):
    assert p["id"] == l["id"]
```

```python
get_rouge(
    [p['title'] for p in pred],
    [l['title'] for l in label]
)
```

```python
get_rouge(
    [p['title'] for p in pred],
    [l['title'] for l in decoded_label]
)
```

```python
for d, l in zip(decoded_label, label):
    if d["title"] != l["title"]:
        print(d["title"], " / ", l["title"])
```
