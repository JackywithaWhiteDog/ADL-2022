```python
%matplotlib inline

import json

from matplotlib import pyplot as plt

plt.style.use('ggplot')
```

```python
with open('../ckpt/mt5_small_5e-4_8/results.json', 'r') as f:
    results = json.load(f)
```

```python
metrics = {
    'Precision': 'p',
    'Recall': 'r',
    'F1 score': 'f'
}
rouges = ['rouge-1', 'rouge-2', 'rouge-l']
epochs = range(1, len(results)+1)

for m, k in metrics.items():
    plt.figure(dpi=100)
    for r in rouges:
        plt.plot(epochs, [elem[r][k] for elem in results])
    plt.title(m)
    plt.legend(rouges)
    plt.xticks(epochs)
    plt.xlabel('Epoch')
    plt.ylabel(m)
    plt.show()
```
