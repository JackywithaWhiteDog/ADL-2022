```python
import random
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
def read_csv(filename):
    idx = []
    data = []
    with open(filename, 'r') as f:
        for line in f:
            idx.append(line[:line.find(',')].strip())
            data.append(line[line.find(',')+1:].strip())
    return {'id': idx, 'source': data}

def read_text(filename):
    with open(filename, 'r') as f:
        data = [line.strip() for line in f.read().split('\n') if line]
    return data
```

```python
sources = {
    'in_domain': '../data/in_domain/test/source.csv',
    'out_of_domain': '../data/out_of_domain/test/source.csv'
}

sources = {
    domain: read_csv(f)
    for domain, f in sources.items()
}
```

```python
files = {
    'in_domain': {
        'target': '../data/in_domain/test/reference.txt',
        'mult-gen': '../models/in_domain/grf-in_domain_anlg_eval/result_ep:test.txt',
        't5': '../ckpt/t5/in_domain/wd1e-1_lr5e-3_batch32_source50_target75/generated_predictions.txt'
    },
    'out_of_domain': {
        'target': '../data/out_of_domain/test/reference.txt',
        'mult-gen': '../models/out_of_domain/grf-out_of_domain_anlg_eval/result_ep:test.txt',
        't5': '../ckpt/t5/out_of_domain/wd1e-1_lr5e-3_batch32_source50_target75/generated_predictions.txt'
    }
}

result = {
    domain: {
        k: read_text(f)
        for k, f in fs.items()
    }
    for domain, fs in files.items()
}
```

```python
domains = ('in_domain', 'out_of_domain')
dfs = {
    domain: pd.DataFrame({**sources[domain], **result[domain]})
    for domain in domains
}
```
